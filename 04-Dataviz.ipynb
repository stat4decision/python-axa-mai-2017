# DataViz

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#on affiche un nuage de points avec deux groupes de points
plt.scatter(np.random.randn(100),np.random.randn(100),s=10,c="red",
            label="Nuage 1")
plt.scatter(np.random.randn(100),np.random.randn(100),s=10,c="blue",
            label="Nuage 2")
plt.title("Mon graphique")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

```python
#on construit un histogramme avec deux échantillons
plt.hist(np.random.randn(10000)+2,bins=50, label="Nuage 1", alpha=0.5,color="red")
plt.hist(np.random.randn(10000),bins=50, label="Nuage 2", alpha=0.5)
plt.annotate("Test",(1,400))
plt.title("Mon histogramme")
plt.xlabel("X")
plt.legend()
```

```python
#on représente trois courbes
x = np.linspace(-2, 2, 100)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
```

```python
#on récupère des données d'Excel
frame_iris=pd.read_excel("iris.xlsx")
```

```python
#on représente un box-plot par espèce
sns.boxplot(y="Espece",x="Larg. Sepales", data=frame_iris)
```

```python
sns.distplot(frame_iris["Larg. Sepales"])
```

```python
sns.pairplot(frame_iris,hue="Espece",diag_kind='kde')
```

```python
#on récupère les données housing.csv
frame_house=pd.read_csv("housing.csv")
```

```python
frame_house.head()
```

```python
#on représente les latitudes et longitudes des points en fixant la taille des points en fonction de la population
#on colorie en fonction de la valeur médiane des logements
plt.scatter("longitude","latitude",data=frame_house, label="Population",
                s=frame_house["population"]/200, c="median_house_value",
           cmap=plt.get_cmap("rainbow"),alpha=0.5)
plt.colorbar()
plt.legend()
```
